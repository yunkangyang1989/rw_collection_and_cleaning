```python
def clean_trending_blp(article):
    """
    method to clean trending links embedded in big league politcs
    :param article: 
    :return: 
    """
    embedded_list = re.findall(r'\n\nTrending:(.*?)\n\n', article)
    for e in embedded_list:
        article = article.replace('\n\nTrending:', '',).replace(e, ' ')
    embedded_list = re.findall(r'\n\nRELATED:(.*?)\n\n', article)
    for e in embedded_list:
        article = article.replace('\n\nRELATED:', '',).replace(e, ' ')
    return article
    
    
s = df[df['media_site']=='Western Journalism Center']['article'].iloc[0]
clean_trending_blp(s)
```

```python
def get_article_contents(url, source):
    try:
        r = requests.get(url)
        data = r.text
        return data
    except:
        try:
            http = source.split("://")[0] + "://"
            prefix = source.split("://")[-1].split('/')[0]
            r = requests.get(http + prefix + url)
            data = r.text
            return data
        except Exception as e:
            return e
```

```python
from gensim.models.doc2vec import Doc2Vec, TaggedDocument
from nltk.tokenize import word_tokenize
```

```python
# prepare for doc2vec
tagged_data = [TaggedDocument(words=word_tokenize(_d.lower()), 
                              tags=[str(i)]) for i, _d in enumerate(text)]
```

```python
# tagged_data[3001]
# If dm=1 means ‘distributed memory’ (PV-DM) and dm =0 means 
# ‘distributed bag of words’ (PV-DBOW). Distributed Memory model 
# preserves the word order in a document whereas Distributed Bag 
# of words just uses the bag of words approach, which doesn’t preserve any word order.
```

```python
max_epochs = 100
vec_size = 300
alpha = 0.025

model = Doc2Vec(vector_size=vec_size,
                alpha=alpha, 
                min_alpha=0.00025,
                min_count=1,
                dm =1)
  
model.build_vocab(tagged_data)

for epoch in range(max_epochs):
    print('iteration {0}'.format(epoch))
    model.train(tagged_data,
                total_examples=model.corpus_count,
                epochs=model.iter)
    # decrease the learning rate
    model.alpha -= 0.0002
    # fix the learning rate, no decay
    model.min_alpha = model.alpha

model.save("d2v.model")
print("Model Saved")
```

```python
def clean_html(article):
    """
    method to return article text from html
    :param article: html
    :return:
    """
    try:
        soup = BeautifulSoup(article, 'html.parser')
        art = ''
        article_text = soup.find_all('p')
        # print text
        for paragraph in article_text:
            # get the text only
            text = paragraph.get_text()
            art += " " + text
        art = art.replace('\n', '')
        return art
    except Exception as e:
        print(e)
        return article
```
